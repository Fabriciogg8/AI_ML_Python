```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Securibank_Default_Dataset.csv", index_col="Customer ID")
df.head()
```

```python
response = df["Default_next_month"]
predictors = df.drop(columns=["Default_next_month"])
```

```python
predictors = pd.get_dummies(predictors)
predictors = predictors.astype(int) #Paso de Booleano a 0 y 1
predictors.head()
```

```python
predictor_train, predictor_test, response_train, response_test = train_test_split(
    predictors, response, test_size=0.15, random_state=99)
```

```python
predictor_train.info()
predictor_test.info()
```

```python
# Histograma de la variable respuesta
plt.figure(figsize=(6,4))
plt.hist(response, bins=2, edgecolor='black')
plt.title("DistribuciÃ³n de Default_next_month")
plt.xlabel("Default_next_month")
plt.ylabel("Cantidad")
plt.show()

# Histogramas de las variables predictoras
for col in predictors.columns:
    plt.figure(figsize=(6,4))
    plt.hist(predictors[col], bins=10,edgecolor='black')
    plt.title(f"DistribuciÃ³n de {col}")
    plt.xlabel(col)
    plt.ylabel("Cantidad")
    plt.show()
```

```python
for col in predictors.columns:
    plt.figure(figsize=(6,4))
    if set(predictors[col].unique()) <= {0,1}:
        bins=2
    else:
        bins=30
    plt.hist(predictors[col], bins=bins, edgecolor='black')
    plt.title(f"DistribuciÃ³n de {col}")
    plt.xlabel(col)
    plt.ylabel("Cantidad")
    plt.show()
```

```python
# Steps
# 1. Train model
# 2. Run model on test data
# 3. Build confusion matrix
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
classifier.fit(predictor_train,response_train)
```

```python
#Step 2
response_predictions = classifier.predict(predictor_test)
```

```python
#Step 3
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix = confusion_matrix(response_test, response_predictions)
print("  TN  FP") #True Negatives, False Positives
print(confusion_matrix)
print("  FN  TP") #False Negatives, True Positives
```

```python
# Otra forma de hacerlo
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Crear y entrenar el modelo con max_depth=1
classifier = DecisionTreeClassifier(max_depth=1)
classifier.fit(predictor_train, response_train)

# 2. Predecir con el modelo ya entrenado
response_predictions = classifier.predict(predictor_test)

# 3. Calcular la precisiÃ³n
accuracy = accuracy_score(response_test, response_predictions)

# 4. Mostrar la precisiÃ³n redondeada a 1 decimal
print(f"La precisiÃ³n con max_depth=1 es: {accuracy:.1f}")
```

```python
print(classification_report(response_test, response_predictions, target_names=["Pay","Defaulted"]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

# 1. Separa variable respuesta
response = df["Default_next_month"]
predictors = df.drop(columns=["Default_next_month"])

# 2. Aplica pd.get_dummies a TODO el dataframe de predictores
predictors = pd.get_dummies(predictors)

# 3. Opcional: conviertes todo a int o float
predictors = predictors.astype(float)  # o int si quieres

# 4. Realiza train/test split
X_train, X_test, y_train, y_test = train_test_split(
    predictors, response, test_size=0.2, random_state=42
)

# 5. Crea el modelo
model = GaussianNB()

# 6. Entrena el modelo
model.fit(X_train, y_train)

# 7. Predice
y_pred = model.predict(X_test)

# 8. EvalÃºa
print(classification_report(y_test, y_pred))
```
